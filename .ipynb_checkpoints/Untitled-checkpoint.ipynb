```python
matchresults = pd.read_csv('groupmatchresults.csv')
matchresults
```

```python
test = matchresults.iloc[16]
url = "https://www.vlr.gg"+test.URL
result = requests.get(url)
soupparser = BeautifulSoup(result.text,'lxml')
url
```

```python
def get_series_info(mapsstats_soup):
    maps_header =mapsstats_soup.find("div",class_="vm-stats-gamesnav noselect")
    series_list = []
    for map_name in maps_header.contents[3:9:2]:
        eachmap_dict = {"Map No.":map_name.text.strip().split()[0],"Map Name":map_name.text.strip().split()[1],
                   "Map ID":map_name["data-game-id"]}
        series_list.append(eachmap_dict)
    return series_list

maps_stats_soup = soupparser.find("div",class_="vm-stats")
seriesinfo = get_series_info(maps_stats_soup)
```

```python
def extract_team_overview(overview_oneteam_soup):
    oneteam_header = overview_oneteam_soup.find("thead")
    oneteam_body = overview_oneteam_soup.find("tbody").find_all("tr")
    oneteam_list= []
    for i in oneteam_body:
        player_stat = []
        for stat in i.find_all("td"):
            if stat.text.strip() != "":
                player_stat.append((stat.text.strip()))

        findteamname = re.compile("[A-Za-z0-9 ]+")
        playerandteam =  re.findall(findteamname, player_stat[0])   
        player_stat[0] = playerandteam[0].strip()
        player_stat.insert(1,playerandteam[1])
        player_stat.insert(2,i.find("img")['title'])
        oneteam_list.append(player_stat)

    colnames = [colname.text.strip() for colname in oneteam_header.find_all("th") if colname.text.strip()!=""]
    colnames_a= ["Player","Team","Agent"]
    colnames_a.extend(colnames)
    colnames_a[-1] = "FK +/-"
    
    df = df = pd.DataFrame(oneteam_list) 
    df.columns =  colnames_a
    return df

def combine_teams_overview(mapsstats_soup,seriesinfo):
    
    statscontainer = mapsstats_soup.find("div",class_="vm-stats-container")
    for game in seriesinfo:
        gamesoup = statscontainer.find("div",attrs={"class":"vm-stats-game","data-game-id":game["Map ID"]})
        bothteams  = gamesoup.find_all("table",class_="wf-table-inset mod-overview")
        c=pd.concat([extract_team_overview(bothteams[0]),(extract_team_overview(bothteams[1]))])
        break
    return c




combine_teams_overview(maps_stats_soup,seriesinfo)
#sen=extract_team_overview(oneteam[0])
```
